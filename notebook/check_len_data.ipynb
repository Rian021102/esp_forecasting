```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/Users/rianrachmanto/miniforge3/project/esp_new.csv')
print(df.head())
```

```python
# make list of names of Well
listwell=df['Well'].unique().tolist()
print(listwell)
```

```python
def plot_voltage(df,listwell):
    for key in listwell:
        plt.figure()
        df[df['Well']==key].plot(x='Date',y='Volt')
        plt.title(key)
        plt.show()
```

```python
plot_voltage(df,listwell)
```
